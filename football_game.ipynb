```python
import turtle
import time, random


wn=turtle.Screen()
wn.title("penaltı")
wn.bgcolor("green")
wn.setup(800,600)
# wn.addshape("kale.gif")
# wn.addshape("kaleci.gif")
# wn.addshape("defance.gif") #burada penceremize oyuncu ekliyoruz


kalem=turtle.Turtle()
kalem.penup()
kalem.setheading(90)
kalem.fd(200)

defance=turtle.Turtle()
defance.shape("triangle") #burada ise oyuncumuza resim ekliyoruz
defance.penup()
defance.speed(6)
defance.goto(0,170)

kaleci=turtle.Turtle()
kaleci.shape("square")
kaleci.penup()
kaleci.speed(0)
kaleci.goto(0,250)
kaleci.turtlesize(6,2)

kale=turtle.Turtle()
kale.shape("classic")
kale.penup()
kale.speed(0)
kale.goto(0,250)
kale.turtlesize(6,2)

top = turtle.Turtle()
top.color("white")
top.shape("circle")
top.penup()
top.speed(6)
top.goto(0,-250)
top.turtlesize(1,1)

# fonksiyon tanımladık funtion defination
def kaleci_saga_git():
    x = kaleci.xcor()
    x = x+15
    kaleci.setx(x)
    
def kaleci_sola_git():
    x = kaleci.xcor()
    x = x-15
    kaleci.setx(x)

def saga_sut():
    x = top.xcor()
    x = x + random.randint(10,200)
    top.goto(x,250)
    time.sleep(1.50)
    top.goto(0,-250)
    
def sola_sut():
    x = top.xcor()
    x = x - random .randint(10,200)
    top.goto(x,250)
    time.sleep(1.50)
    top.goto(0,-250)
    
def top_saga_git():
    x = top.xcor()
    x = x+15
    top.setx(x)

def top_sola_git():
    x = top.xcor()
    x = x-15
    top.setx(x)
    
    
wn.listen()
wn.onkey(kaleci_saga_git, "a")
wn.onkey(kaleci_sola_git, "d")

wn.onkey(top_saga_git, "Right")
wn.onkey(top_sola_git, "Left")
wn.onkey(saga_sut, "3")
wn.onkey(sola_sut,"1")

#burada sonsuz bir döngü yapıyoruz
while True : #while olduğunca demek true doğru demek sonuc olarak doğru olduğu sürece devam et demek 
    
    defance_angle = defance.towards(top)#oyuncumuzun yönünü değiştirdik
    defance.setheading(defance_angle)
    defance.fd(2)
    
    if defance.ycor() < -240: #eger defansın y kordinatı alt kenarın altına gelirse iki nokta önemli
        defance.goto(0,150) #burada başlangıç noktasına gidiyo burası if in gövde olduğu için girinti bıkaktık (indent)
    elif defance.distance(top) < 10:
        turtle.goto(0,150)
        turtle.up()
        turtle.ht()
        turtle.color("red")
        turtle.write("Game Over", font = ("Corier", 24, "bold"))
    


wn.mainloop()
```
